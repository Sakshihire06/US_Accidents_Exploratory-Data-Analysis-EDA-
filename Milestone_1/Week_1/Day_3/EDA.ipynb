```python
missing_values=accidental_df.isnull().sum()
missing_percent=(missing_values/len(accidental_df)*100 )
missing=pd.DataFrame ({'Missing Values':missing_values,
                      'Missing percent':missing_percent})
display( missing)
```

```python
missing=missing[missing['Missing Values']>0].sort_values(by='Missing Values',ascending=False)
print('Colums with missing values in descending order')
print(missing)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(accidental_df[missing.index].isnull(),cbar=False,cmap='viridis')
plt.title('Missing values heatmap')
plt.show()
```
